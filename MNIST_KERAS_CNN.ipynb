## Chargement des donnГ©es et transformation

Nous allons travailler sur la base de donnГ©es MNIST qui contient 60000 images en niveaux de grille de rГ©solution 28x28, reprГ©sentant les 10 chiffres de 0 Г  9, ainsi qu'un jeu de test de 10000 images. Tout d'abord, chargeons ce jeu de donnГ©es.

```python
import tensorflow as tf
print("TensorFlow version:", tf.__version__)
```

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import to_categorical
import matplotlib
import matplotlib.pyplot as plt
import scipy
import scipy.ndimage
import numpy as np
from keras import backend as K
import keras

(X_train_base, y_train_base), (X_test_base, y_test_base) = mnist.load_data()
```

Visualisons quelques donnГ©es.   

```python
def plot_10_by_10_images(images):
    """ Plot 100 MNIST images in a 10 by 10 table. Note that we crop
    the images so that they appear reasonably close together.  The
    image is post-processed to give the appearance of being continued."""
    fig = plt.figure()
    images = [image[3:25, 3:25] for image in images]
    #image = np.concatenate(images, axis=1)
    for x in range(10):
        for y in range(10):
            ax = fig.add_subplot(10, 10, 10*y+x+1)
            ax.matshow(images[10*y+x+1], cmap = matplotlib.cm.binary)
            plt.xticks(np.array([]))
            plt.yticks(np.array([]))
    plt.show()

plot_10_by_10_images(X_train_base)
```

Maintenant que nous avons chargГ© les donnГ©es, nous allons modifier la dimension de matrices, afin de les mettre sous une forme qui pourra ГӘtre traitГ©e par nos rГ©seaux de neurones.

```python
subset=10000 #size
nb_classes=10
X_train = X_train_base[:subset].reshape(subset, 784)
X_test = X_test_base.reshape(10000, 784)
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
X_train /= 255
X_test /= 255
y_train = to_categorical(y_train_base[:subset], nb_classes)
y_test = to_categorical(y_test_base, nb_classes)
```

Nous avons maintenant des vecteurs Г  la place des images.

```python
print(X_train_base.shape)
print(X_train.shape)
```

## Construire un premier rГ©seau de neurones

Construisons notre premier rГ©seau de neurones.

Pour cela, nous allons crГ©er un modГЁle Keras en utilisant l'api Sequential:
* __model = Sequential()__

Puis utiliser les mГ©thodes suivantes de Keras pour ajouter des couches Г  ce modГЁle :

* __model.add(Dense(nbhiden, [input_shape=(nbinput,1)], [activation='activationfunction']))__ : ajoute une couche cachГ©e au rГ©seau de neurones. nbhidden est le nombre de neurones de la couche. [input_shape=(nbinput,)] correspond Г  la dimension de l'entrГ©e, [activation='activationfunction'] est la fonction d'activation
* __model.add(Dropout(value))__ : applique un dropout Г  la couche, pour Г©viter le surapprentissage

```python
model = Sequential()
model.add(Input(shape=(784,)))
model.add(Dense(12,activation='sigmoid'))
model.add(Dense(12,activation='sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(nb_classes, activation='softmax'))

sgd = SGD(learning_rate=0.1, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],  optimizer=sgd)

model.summary()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.optimizers import SGD

nb_classes = 10   # Г  dГ©finir selon votre problГЁme

model = Sequential()
model.add(Input(shape=(784,)))
model.add(Dense(12, activation='sigmoid'))
model.add(Dense(12, activation='sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(nb_classes, activation='softmax'))

sgd = SGD(learning_rate=0.1, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=sgd)

model.summary()
```

CrГ©ation du modГЁle sГ©quentiel

PremiГЁre couche cachГ©e

**model.add(Dense(12, input_shape=(784,), activation='sigmoid')**)


вҖў	12 neurones : c'est le nombre de neurones de la couche cachГ©e. Assez faible, mais peut suffire pour des problГЁmes simples.

вҖў	input_shape=(784,) : spГ©cifie que l'entrГ©e est un vecteur de 784 dimension 784 (par exemple des images 28Г—28 aplaties).  valeurs (chaque image aplatie). Cette information n'est nГ©cessaire que pour la premiГЁre couche.

вҖў	Fonction d'activation sigmoid : produit des valeurs entre 0 et 1. Historiquement utilisГ©e, mais aujourd'hui on prГ©fГЁre relu pour les couches cachГ©es car elle Г©vite le problГЁme de disparition du gradient et accГ©lГЁre la convergence.


** DeuxiГЁme couche cachГ©e**

**model.add(Dense(12, activation='sigmoid'))**
Une deuxiГЁme couche avec 12 neurones, toujours avec sigmoid. Cela donne un rГ©seau de profondeur modeste.

**Dropout**

**model.add(Dropout(0.5))**
вҖў	Dropout(0.5) : dГ©sactive alГ©atoirement 50% des neurones pendant l'entraГ®nement Г  chaque itГ©ration. C'est une technique de rГ©gularisation trГЁs efficace pour rГ©duire le surapprentissage. Ici, il est placГ© aprГЁs la deuxiГЁme couche cachГ©e. Il pourrait aussi ГӘtre placГ© aprГЁs la premiГЁre.


**Couche de sortie**

**model.add(Dense(nb_classes, activation='softmax'))**

вҖў	nb_classes : doit ГӘtre dГ©fini avant (par exemple nb_classes = 10 pour MNIST/Fashion-MNIST). Cette couche donne une distribution de probabilitГ© sur les classes.

вҖў	Activation softmax : transforme les scores en probabilitГ©s (somme = 1). C'est la fonction standard pour la classification multi-classe.
** Optimiseur SGD**

**sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)**


вҖў	lr=0.1 : taux d'apprentissage assez Г©levГ© (attention Г  la stabilitГ©).

вҖў	decay=1e-6 : rГ©duit progressivement le taux d'apprentissage au fil des itГ©rations (learning rate decay).

вҖў	momentum=0.9 : accГ©lГЁre la descente de gradient en ajoutant une fraction de la mise Г  jour prГ©cГ©dente.

вҖў	nesterov=True : utilise la version Nesterov du momentum (souvent plus performante).

C'est un optimiseur SGD classique avec momentum. Pour beaucoup de problГЁmes, Adam est plus simple Г  rГ©gler et converge souvent plus vite, mais SGD bien paramГ©trГ© peut donner d'excellents rГ©sultats.



**Compilation**

**model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=sgd)**

вҖў loss='categorical_crossentropy' : adaptГ©e aux sorties one-hot encoded.

вҖў metrics=['accuracy'] : on suit la prГ©cision pendant l'entraГ®nement.

вҖў optimizer=sgd : l'optimiseur dГ©fini prГ©cГ©demment

Ensuite, nous allons lancer l'apprentissage des paramГЁtres.

```python
batch_size = 2048
epochs=3000
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs,  verbose=1, validation_data=(X_test, y_test))
```

## CNN : rГ©seaux de neurones convolutionnels

Nous allons maintenant implГ©menter un rГ©seau de neurones convolutionnel.

Pour cet exercice, vous allez avoir besoin des mГ©thodes Keras suivantes, en plus de celles dГ©jГ  vues prГ©cГ©demment :

* __model.add(Conv2D(filters, kernel_size=(size, size), [activation='activation_function'], [input_shape=input_shape]))__ : ajoute une couche de convolution au rГ©seau de neurones. filters est le nombre de filtres de convolution. kernel_size est la taille des filtres de convolution.
* __model.add(MaxPooling2D(pool_size=(psize, psize)))__ : ajoute une couche de max pooling. pool_size est la taille de la fenГӘtre de pooling.

```python
img_rows, img_cols = 28, 28
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train[:subset].reshape(x_train[:subset].shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
y_train = keras.utils.to_categorical(y_train[:subset], nb_classes)
y_test = keras.utils.to_categorical(y_test, nb_classes)
```

```python
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
model = Sequential()
model.add(Conv2D(4, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(10, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(nb_classes, activation='softmax'))

sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],  optimizer=sgd)

model.summary()
```

```python
batch_size = 256
epochs=20
model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs,  verbose=1, validation_data=(x_test, y_test))
```

```python
def plot_mnist_digit(image):
    """ Plot a single MNIST image."""
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.matshow(image, cmap = matplotlib.cm.binary)
    plt.xticks(np.array([]))
    plt.yticks(np.array([]))
    plt.show()
loss,acc = model.evaluate(x_test, y_test,  verbose=0)
index=800
print('The accuracy on the test set is ',(acc*100),'%')
plot_mnist_digit(X_test_base[index])
cl=model.predict_classes(x_test[index].reshape((1,28,28,1)))


print("le chiffre reconnu est: ", cl[0])
print("le chiffre Г  reconnaitre  est: ", np.argmax(y_test[index]))
```

## Bonus : Auto encodeur

L'auto-encodeur est un rГ©seau de neurones qui compresse puis dГ©compresse l'information. On l'entraГ®ne en lui demandant de retrouver en sortie la mГӘme image que celle qu'il avait en entrГ©e. Ici, l'information en entrГ©e est en dimension 784 (28x28), et l'auto-encodeur va la compresser en dimension 2.

```python
encoding_dim = 2

model = Sequential()
model.add(Dense(encoding_dim, input_shape=(784,),activation='relu'))

model.add(Dense(784, activation='sigmoid'))

sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='binary_crossentropy',  optimizer=sgd)

model.summary()
```

```python
(X_train, _), (X_test, _) = mnist.load_data()
subset=50000 #size
X_train = X_train[:subset].reshape(subset, 784)
X_test = X_test.reshape(10000, 784)
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")
X_train /= 255
X_test /= 255
```

```python
model.fit(X_train, X_train,
                epochs=50,
                batch_size=256,
                shuffle=True,
                validation_data=(X_test, X_test))
```

```python
decoded_imgs = model.predict(X_test)
```

Affichons quelques images pour voir comment se comporte notre auto-encodeur.

```python
import matplotlib.pyplot as plt

n = 10  # how many digits we will display
plt.figure(figsize=(20, 4))
for i in range(n):
    # display original
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(X_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display reconstruction
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```
